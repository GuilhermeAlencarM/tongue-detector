```python
import pandas as pd
import random

# Lendo as frases do arquivo de texto
with open('/content/dataset.txt', 'r', encoding='utf-8') as file:
    frases = [line.strip() for line in file]

# Criando o DataFrame
data = pd.DataFrame(columns=['frase', 'sentimento'])

# Populando o DataFrame com 1001 registros
for _ in range(1001):
    frase = random.choice(frases)
    data = data.append({'frase': frase, 'sentimento': 'felicidade'}, ignore_index=True)

# Salvando o DataFrame em um arquivo CSV
data.to_csv('dataset_felicidade.csv', index=False)
```

```python
pd.read_csv("/content/dataset_felicidade.csv")
data
```

```python
import seaborn as sns
sns.countplot(data, x="sentimento")
```

```python
import pandas as pd

# Carregar os trs conjuntos de dados
dataset_raiva = pd.read_csv('dataset_raiva.csv')
dataset_tristeza = pd.read_csv('dataset_tristeza.csv')
dataset_felicidade = pd.read_csv('dataset_felicidade.csv')

# Mesclar os conjuntos de dados
merged_dataset = pd.concat([dataset_raiva, dataset_tristeza, dataset_felicidade])

# Salvar o conjunto de dados mesclado em um novo arquivo CSV
merged_dataset.to_csv('merged_dataset.csv', index=False)

print("Conjunto de dados mesclado criado com sucesso!")
```
